```python
# import library
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from apyori import apriori
```

```python
# membaca data
dt = pd.read_csv('data/order_detail_preprocessing.csv', header=None)
```

```python
# melihat 5 data teratas
dt.head()
```

```python
rows = []
for i in range (0, len(dt)):
    # mengambil data transaksi
    rows.append(dt.iloc[i,1].split(','))
```

```python
rows
```

```python
# training apriori
min_lift_dt = 2
support_dt = 0.002
conf_rule = 1
asc_rules = apriori(rows, min_support = support_dt, min_confidence = conf_rule, min_lift = min_lift_dt, min_length = 3)
asc_rules_result = list(asc_rules)
```

```python
print(len(asc_rules_result))
```

```python
asc_rules_result
```

```python
MB = list(asc_rules_result)
listRules = [list(MB[i][0]) for i in range(0,len(MB))]
for i in range(0, len(MB)):
    MC = list(MB[i][2])
    listRules[i].append({'conf':MC[0][2]})
    listRules[i].append({'lift':MC[0][3]})
```
