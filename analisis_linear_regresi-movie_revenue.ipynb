```python
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt
import math
import sklearn

general_data = pd.read_csv("/Users/hendri/Downloads/movies_data_v2.csv")
general_data.head(5)
```

```python
general_data.plot(x='x', y='y', style='o')
plt.title('pengaruh biaya produksi terhadap revenue')
plt.xlabel('Biaya produksi')
plt.ylabel('revenue')
plt.show()
```

```python
x = general_data.iloc[:, :-1].values
y = general_data.iloc[:, 1].values
x
```

```python
from sklearn.model_selection import train_test_split 
X_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)

from sklearn.linear_model import LinearRegression
regressor = LinearRegression()

#X_train = X_train.reshape(1, -1)
#y_train = y_train.reshape(1, -1)
#x_test = x_test.reshape(1, -1)
#X_train = X_train.fillna(X_train.mean())

regressor.fit(X_train, y_train)
print(regressor.intercept_)
print(regressor.coef_)

plt.scatter(X_train, y_train, color = 'red')
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.title('pengaruh biaya produksi terhadap revenue')
plt.xlabel('Biaya produksi')
plt.ylabel('revenue')
plt.show()
```
